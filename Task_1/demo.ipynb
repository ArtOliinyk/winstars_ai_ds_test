# Task 1 Demo: MNIST Classifier Comparison

This notebook demonstrates how the Task 1 solution works and checks edge cases.

- Models included: Random Forest (RF), Feedforward NN (NN), Convolutional NN (CNN)
- It uses `main.MnistClassifier` wrapper from `Task_1/main.py`.
- Edge cases: invalid input shape, empty input, value range check.

```python
from main import MnistClassifier
from tensorflow.keras.datasets import mnist
from sklearn.metrics import accuracy_score
import numpy as np
import warnings
warnings.filterwarnings('ignore')

(X_train, y_train), (X_test, y_test) = mnist.load_data()
print('MNIST data shapes:', X_train.shape, X_test.shape)
```

```python
def evaluate(algorithm):
    model = MnistClassifier(algorithm=algorithm)
    model.train(X_train[:1000], y_train[:1000])  # quick demo training subset
    y_pred = model.predict(X_test[:1000])
    acc = accuracy_score(y_test[:1000], y_pred)
    return acc

for algo in ['rf', 'nn', 'cnn']:
    print(f'{algo.upper()} accuracy (subset):', evaluate(algo))
```

## Edge Case Testing

```python
from main import RandomForestMnistClassifier, NeuralNetworkMnistClassifier, CNNMnistClassifier

rf = RandomForestMnistClassifier()
nn = NeuralNetworkMnistClassifier(epochs=1)
cnn = CNNMnistClassifier(epochs=1)

bad_inputs = [
    np.zeros((10, 20, 20)),  # wrong size 
    np.zeros((10, 28, 28, 1)),  # wrong dimensions for RF/NN
    np.array([]).reshape(0, 28, 28)  # empty set
]

for z in bad_inputs:
    for cls in [rf, nn, cnn]:
        try:
            cls.predict(z)
            print('Unexpected success for', cls.__class__.__name__, 'with input', z.shape)
        except Exception as e:
            print('Expected exception for', cls.__class__.__name__, 'with input', z.shape, ':', type(e).__name__, e)
```

## Results
- The selection of classifier is by `MnistClassifier(algorithm='rf'|'nn'|'cnn')`.
- Accuracy shown is on a small subset for quick demonstration.
- Edge cases should raise `ValueError` for malformed shapes and handle empty input with clear errors.
